```python
def switching_countries(df):
    df["country"].replace({"None": None}, inplace=True)
    df["country"].replace({"U.A.E.": "United Arab Emirates"}, inplace=True)
    df["country"].replace({"U.K.": "United Kingdom"}, inplace=True)
    df["country"].replace({"England": "United Kingdom"}, inplace=True)
    df["country"].replace({"Wales": "United Kingdom"}, inplace=True)
    df["country"].replace({"U.S.": "United States"}, inplace=True)
```

```python
def merge_enhanced_data(original_df, toadd_df, column_to_improve, column_to_drop):
    original_df["SearchName"] = original_df["name"].apply(lambda full_name: "_".join(full_name.strip().split()))
    enhanced_df = pd.merge(original_df, toadd_df, how="left", on="SearchName")
    for x in enhanced_df.index:
        if pd.isna(enhanced_df.loc[x, column_to_improve]):
            enhanced_df.loc[x, column_to_improve] = enhanced_df.loc[x, column_to_drop]
    enhanced_df.drop(columns=["SearchName", column_to_drop], axis=1, inplace=True)
    enhanced_df.rename(columns={column_to_drop: column_to_improve}, inplace=True)
    enhanced_df = enhanced_df[['id', 'name', 'age', 'gender', 'country', 'image', 'Sector', 'Company', 'Worth(BUSD)']]
    switching_countries(enhanced_df)
    print(f"Data was improved by adding new {column_to_improve} values")
    return enhanced_df
```

```python
final = merge_enhanced_data(original, dftoadd, "country", "Nationality")
```

```python
def api_search_list (df_clean, df_CC, tolist=False):
    df_clean = df_clean[["id", "name", "gender", "country"]].copy()
    fn = df_clean[df_clean["gender"].isnull()].copy()
    fn["first_name"] = fn["name"].apply(lambda full_name: full_name.strip().split()[0])
    merged_df = pd.merge(fn, df_CC, how="left", on="country")
    merged_df["accuracy"] = 0
    if tolist == True:
        return merged_df.values.tolist()
    else:
        return merged_df
```

```python
search_gender = api_search_list(final, country_relation, tolist=True)
search_gender_df = api_search_list(final, country_relation, tolist=False)
```

```python
import requests
def api_search(api_search_lst, todf=False):
    for x in range(5):
        """
        [3808, 'Bernard Arnault', nan, 'France', 'Bernard', 'FR', 0]
        [id, 'Full Name', Gender, 'Country', 'First Name', 'Code', Accuracy]
        """
        name = api_search_lst[x][4]
        code =  api_search_lst[x][5]
        
        r = requests.get(f"https://api.genderize.io?name={name}&country_id={code}")
        
        api_search_lst[x][2] = r.json()["gender"].capitalize()
        api_search_lst[x][6] = r.json()["probability"]
        
    if todf == True:
        cols = ['id', 'name', 'new_gender', 'country', 'first_name', 'code', 'accuracy']
        fulldf = pd.DataFrame(api_search_lst, columns=cols)
        df = fulldf[["id","new_gender", "accuracy"]].copy()
        return df
    else:
        return api_search_lst
```

```python
filled_list = api_search(search_gender, todf=True)
filled_list.head()
```

```python
def merge_enhanced_gender(clean_df, toadddf):
    data_enhanced_gender = pd.merge(clean_df, toadddf, how="left", on="id")
    for x in data_enhanced_gender.index:
        if pd.isna(data_enhanced_gender.loc[x, "gender"]):
            data_enhanced_gender.loc[x, "gender"] = data_enhanced_gender.loc[x, "new_gender"]
    data_enhanced_gender.drop(columns=["new_gender"], axis=1, inplace=True)
    return data_enhanced_gender
```
