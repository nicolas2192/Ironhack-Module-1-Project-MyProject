```python
bar1_vals = [10,20]
bar1_labs = ["azul", "verde"]
bar2_vals = [5,15]
bar2_labs = ["rojo", "amarillo"]

plt.figure(figsize=(12,6))
plt.suptitle("Three random bar charts")

plt.subplot(1,2,1)
plt.title("bar 1")
plt.bar(bar1_labs, bar1_vals)

plt.subplot(1,2,2)
plt.title("bar 2")
plt.bar(bar2_labs, bar2_vals)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

width = 0.8

highPower   = [2,4,6,8,10]
lowPower    = [1,2,3,4,5]

indices = np.arange(len(highPower))

plt.bar(indices, highPower, width=width, 
        color='b', label='Max Power in mW')
plt.bar([i+0.25*width for i in indices], lowPower, 
        width=0.5*width, color='r', alpha=0.5, label='Min Power in mW')

plt.xticks(indices+width/2., 
           ['T{}'.format(i) for i in range(len(highPower))] )

plt.legend()

plt.show()
```
